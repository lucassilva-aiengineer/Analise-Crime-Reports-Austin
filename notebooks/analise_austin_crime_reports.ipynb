```python
df.columns
```

```python
# 1. Idade

idades_validas = df[df["Vict Age"] > 0]
```

```python
idades_validas["Vict Age"].hist(color= "darkblue")

plt.show()
```

```python
from typing import List
```

```python
# Eu preciso criar quantis 

idades: List[int] = []
```

```python
for idade in list(idades_validas["Vict Age"]):
    if idade <= 20:
        idades.append(20)

    elif idade <= 40: 
        idades.append(40)

    elif idade <= 60:
        idades.append(60)

    elif idade <= 80:
        idades.append(80)

    else:
        idades.append(100)
```

```python
from collections import Counter
```

```python
contagem_idades = Counter(idades)
```

```python
idades, valores = zip(*contagem_idades.items())
```

```python
idades
```

```python
rotulos = np.array(["20 até 40", "40 até 60", "60 até 80", "até 20", "mais de 100" ])
```

```python
plt.bar(rotulos, valores,
       color= "lightblue",
       edgecolor= "black")

plt.title("Idade das vítimas")
plt.grid(True, linestyle= "--")

plt.show()
```
